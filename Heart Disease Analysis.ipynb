# No of People suffering from heart disease

```python
heart_disease_dataset_copy['target'].value_counts()
```

```python
target_counts=heart_disease_dataset_copy['target'].value_counts()
plt.bar(target_counts.index,target_counts.values)
plt.xlabel('Target')
plt.xticks([0,1],['Healthy Person','Diseased Person'])
plt.ylabel('Target Counts')
plt.title('Distribution of Target Variable')
plt.show()
```

# Total Number of Male and Female

```python
heart_disease_dataset_copy['sex'].value_counts()
```

```python
sex_counts=heart_disease_dataset_copy['sex'].value_counts()
plt.bar(sex_counts.index,sex_counts.values)
plt.xlabel('Sex')
plt.xticks([0,1],['Female','Male'])
plt.ylabel('Sex Counts')
plt.title('Distribution of Sex Counts')
plt.show()
```

# Age Distribution In The Dataset

```python
plt.hist(heart_disease_dataset_copy['age'],bins=20)
plt.show()
```

# Types of Chest Pain

```python
cp_counts=heart_disease_dataset_copy['cp'].value_counts()
plt.bar(cp_counts.index,cp_counts.values)
plt.xlabel('Chest Pain Type')
plt.xticks([0,1,2,3],['Typical Angina','Atypical Angina','Non-Anginal Pain','Asymptomatic'])
plt.ylabel('Chest Pain Counts')
plt.title('Distribution of Chest Pain')
plt.show()
```

# Chest Pain Distribution

```python

sns.countplot(x='cp',hue='target',data=heart_disease_dataset_copy)
plt.xticks([0,1,2,3],['Typical Angina','Atypical Angina','Non-Anginal Pain','Asymptomatic'])
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```

# Fasting Blood Sugar Distribution

```python
sns.countplot(x='fbs',hue='target',data=heart_disease_dataset_copy)
plt.xticks([0,1],['False','True'])
plt.legend(labels=['Disease','No-Disease'])
plt.show()
```

# Resting Blood Sugar Distribution

```python
heart_disease_dataset_copy['trestbps'].hist(grid=False)
plt.show()
```

# Resting Blood Pressure As Per Sex Column

```python
g=sns.FacetGrid(heart_disease_dataset_copy,hue='sex',aspect=6)
g.map(sns.kdeplot,'trestbps',shade=True)
plt.legend(labels=['Male','Female'])
plt.show()
```

```python
print('Min Age: ',heart_disease_dataset_copy['age'].min())
print('Median Age: ',heart_disease_dataset_copy['age'].median())
print('Max Age: ',heart_disease_dataset_copy['age'].max())
```

```python
young_age_people=heart_disease_dataset_copy[(heart_disease_dataset_copy['age']>=29) & (heart_disease_dataset_copy['age']<40)]
middle_age_people=heart_disease_dataset_copy[(heart_disease_dataset_copy['age']>=40) & (heart_disease_dataset_copy['age']<55)]
elderly_age_people=heart_disease_dataset_copy[(heart_disease_dataset_copy['age']>55)]
print('Young Ages =',len(young_age_people))
print('Middle Ages =',len(middle_age_people))
print('Elderly Ages =',len(elderly_age_people))
```

```python
plt.figure()
sns.barplot(x=['young_age_people','middle_age_people','elderly_age_people'],y=[len(young_age_people),len(middle_age_people),len(elderly_age_people)])
plt.xlabel('Age Category')
plt.ylabel('Ages')
plt.title('Age Distribution')
plt.show()
```
